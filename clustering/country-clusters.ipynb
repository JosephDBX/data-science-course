## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# sns.set()
```

```python
data = pd.read_csv('./country-clusters.csv', index_col='Country')
x_scaled = data.copy()
x_scaled = x_scaled.drop(['Language'], axis=1)
x_scaled
```

## Plot the data

```python
sns.clustermap(x_scaled, cmap='mako')
```

- Canada and USA looks like the same in term on longitude and latitude, their can be a cluster, based in the heatmap
- Germany, France and UK can be a cluster
- Australia remain alone in its own cluster
